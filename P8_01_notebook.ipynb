## Brève description

La Coupe du monde de football 2022 est la 22e édition de la Coupe du monde de football, compétition organisée par la FIFA et qui réunit les meilleures sélections nationales. Elle se déroulera au Qatar du 21 novembre au 18 décembre 2021, jour de la fête nationale du Qatar et une semaine avant Noël, avec une estimation du marché télévisuel potentiel à 3,2 milliards de téléspectateurs. La période inédite à laquelle se déroulera le tournoi est liée au climat du Qatar et aux trop fortes chaleurs y régnant, particulièrement aux dates où se déroule habituellement la compétition (mai-juin-juillet).

L'étude a pour objectif de proposer une vue d'ensemble afin de pouvoir segmenter toutes les éqipes participantes de l'édition 2022 de la Coupe du Monde de Football, en se focalisant sur plusieurs aspects sportifs et économiques

## Les enjeux :

### Un enjeu sportif 
Avec à la clé, un rayonnement de 4 ans sur la scène footballistique mondiale, des retombées économiques au niveau marketing et sponsoring
### Un enjeu politique
Le pays hôte pour qui l'image sera améliorée le temps de la compétition voir après
### Un enjeu économique
Création d'emplois, un retour sur investissement possible avec les recettes

### A) Analyse en Composantes Principales

```python
import pandas as pd
import numpy as np
from sklearn import decomposition
from sklearn import preprocessing
```

```python
from functions import display_circles, display_scree_plot, display_factorial_planes
```

```python
# selection des colonnes à prendre en compte dans l'ACP
worldcup2022_acp = worldcup2022.copy()
```

```python
worldcup2022_acp=worldcup2022_acp[['Teams','PIB/hab','FIFA Number points','pop_2020','AVG Age','Market value €','Number CL Winners','Number Players abroad']]
```

```python
worldcup2022_acp
```

```python
# préparation des données pour l'ACP
#worldcup2022_acp = worldcup2022_acp.fillna(worldcup2022_acp.median()) # Il est fréquent de remplacer les valeurs inconnues par la mediane de la variable
worldcup2022_acp.index=worldcup2022_acp.Teams
X= worldcup2022_acp.drop(['Teams'],axis=1)
X= X.values
names = worldcup2022_acp.index # ou data.index pour avoir les intitulés
features = worldcup2022_acp.columns[1:]
```

```python
features
```

```python
# Centrage et Réduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
# choix du nombre de composantes à calculer
n_comp = 3
```

```python
# Calcul des composantes principales
acp = decomposition.PCA(n_components=n_comp)
acp.fit(X_scaled)
```

```python
# Eboulis des valeurs propres
display_scree_plot(acp)
```

```python
#Selon le pourcentage d'inertie, on peut choisir comme nombre de composants : 3
```

```python
# Cercle des corrélations
pcs = acp.components_
```

```python
display_circles(pcs, n_comp, acp, [(0,1),(0,2),(1,2)], labels = np.array(features))
```

```python
# Projection des individus
X_projected = acp.transform(X_scaled)
display_factorial_planes(X_projected, n_comp, acp, [(0,1),(0,2),(1,2)], labels = np.array(names))

plt.show()
```

```python
def plot_dendrogram(Z, names):
    plt.figure(figsize=(15,10))
    plt.title('Dendrogramme / Clustering Hierarchique')
    plt.xlabel('Equipes nationales')
    dendrogram(Z, labels = names, orientation = "top")
    plt.show()
```

### B) K-Means Clustering

```python
# Comparaison des clusters 
groupes = pd.DataFrame({"Cluster": clusters, "Teams": names})
#print(groupes)
groupes.sort_values(by=['Cluster'])
```

```python
# Transform the scaled data to the new PCA space
X_reduced = acp.transform(X_scaled)
display_factorial_planes(X_reduced, 6, acp, [(0,1)], illustrative_var = clusters, alpha = 0.8)
```

```python
features
```

```python
from scipy.cluster.hierarchy import centroid, fcluster
from scipy.spatial.distance import pdist
y = pdist(X_scaled)
# centroides
centroid(y)
```

```python
from sklearn import cluster
kmeans = cluster.KMeans(n_clusters=6)
kmeans.fit(X_scaled)
# les centroïdes des groupes et leurs coordonnées dans chacune des dimensions.
print(kmeans.cluster_centers_)
```

```python
# Affichage des positions des centres de classes
plt.figure()
centroids = kmeans.cluster_centers_
centroids_projected = acp.transform(centroids)
plt.scatter(centroids_projected[:,0],centroids_projected[:,1], label = 'True position')
plt.title("Projection des {} centres sur le 1e plan factoriel".format(len(centroids)))
for cluster, x, y in zip(set(clusters.tolist()), centroids_projected[:, 0], centroids_projected[:, 1]):
    plt.annotate(
       cluster,
     xy=(x, y), xytext=(-3, 3),
     textcoords='offset points', ha='right', va='bottom')
plt.show()
```
